```python
import re
def clear_special_char(content):
    '''
    正则处理特殊字符
    参数 content:原文本
    return: 清除后的文本
    '''
    s = re.sub(r"</?(.+?)>|&nbsp;|\t|\r", "", content)
    s = re.sub(r"\n", " ", s)
    s = re.sub('[^\u4e00-\u9fa5^a-z^A-Z^0-9]', '', s)

    return s
```

```python
from openpyxl import load_workbook
open_wb=load_workbook(r'C:\Users\Administrator\Desktop\demo_excel1.xlsx') #文件需要提前创建
sheet=open_wb['Sheet1']
sheet.append(['编号', '游戏ID', '游戏名称']) #设置表头
ID = '编号'
GameId = '游戏ID'
GameName = '游戏名称'
i=0
for game in info:
    i+=1
    ID = i
    GameId = game['游戏ID']
    content =game['游戏名称'] 
    GameName = clear_special_char(content)
    sheet.append([ID, GameId, GameName])
open_wb.save(r'C:\Users\Administrator\Desktop\demo_excel1.xlsx')
```
